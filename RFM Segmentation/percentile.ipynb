```python
# Sort customer based on a measure
# Break customers into a pre-define number of groups of equal size
# Assign a label to each group

import pandas as pd

data = {'CustomerId': [0, 1, 2, 3, 4, 5, 6, 7],
        'Spend':[137, 335, 172, 355, 303, 233, 244, 229]}

dataframe = pd.DataFrame(data=data)

dataframe.head(10)
```

```python
spend_quartile = pd.qcut(dataframe['Spend'], q=4, labels=range(1, 5))
dataframe['spend_quartile'] = spend_quartile
dataframe = dataframe.sort_values('spend_quartile')
dataframe.head(10)
```

```python
spend_quartile = pd.qcut(dataframe['Spend'], q=4, labels=range(4, 0, -1))
dataframe['spend_quartile'] = spend_quartile
dataframe = dataframe.sort_values('spend_quartile')
dataframe.head(10)
```

```python
labels = ['Active', 'Lapsed', 'Inactive', 'Churned']
spend_quartile = pd.qcut(dataframe['Spend'], q=4, labels=labels)
dataframe['spend_quartile'] = spend_quartile
dataframe = dataframe.sort_values('spend_quartile')
dataframe.head(10)
```
